```python
import pandas as pd 
vine_df = pd.read_csv("data/vine_table.csv")
vine_df.head()
```

```python
#filter to total votes > 20
total_votes_df = vine_df[vine_df['total_votes']>20]
total_votes_df.head()
```

```python
# create df where helpful votes divided/total votes >50%
helpful_votes_df = total_votes_df[total_votes_df["helpful_votes"]/total_votes_df["total_votes"]>= 0.5]
helpful_votes_df.head()
```

```python
#filter for vine program
vine_program_df = helpful_votes_df[helpful_votes_df["vine"]=="Y"]
vine_program_df
```

```python
#unpaid, not part of vine program
unpaid_df = helpful_votes_df[helpful_votes_df["vine"]=="N"]
unpaid_df
```

```python
#number of five star reviews for both unpaid and paid category and total reviews
paid_five_stars = len(vine_program_df[vine_program_df['star_rating']== 5])
total_paid_reviews = len(vine_program_df)

unpaid_five_stars = len(unpaid_df[unpaid_df['star_rating']==5])
total_unpaid_reviews = len(unpaid_df)
```

```python
#paid five star review percentage
paid_percentage = (paid_five_stars/total_paid_reviews)*100
paid_percentage
```

```python
#unpaid five star percentage
unpaid_percentage = (unpaid_five_stars/total_unpaid_reviews)*100
unpaid_percentage
```
